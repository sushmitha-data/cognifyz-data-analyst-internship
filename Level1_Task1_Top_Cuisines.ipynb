```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\SUSHMITHA RAJU\Downloads\40811eyack.com.MAIL_xsbsxxypt8dh6!App\Dataset._csv.csv")

# LEVEL 1 - TASK 1: Top Cuisines
top_cuisines = df['Cuisines'].value_counts().head(3)
print("=" * 50)
print("TOP 3 MOST COMMON CUISINES")
print("=" * 50)

total = len(df)
for cuisine, count in top_cuisines.items():
    percentage = (count / total) * 100
    print(f"{cuisine}: {count} restaurants ({percentage:.2f}%")
plt.figure(figsize=(10, 6))
plt.bar(top_cuisines.index, top_cuisines.values,
        color=['tomato', 'steelblue', 'green'])
plt.title("Top 3 Most Common Cuisines", fontsize=16, fontweight='bold')
plt.xlabel("Cuisine", fontsize=12)
plt.ylabel("Number of Restaurants", fontsize=12)
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig("Level1_Task1_Top_Cuisines.png")
plt.show()
print("Chart saved!")
```
