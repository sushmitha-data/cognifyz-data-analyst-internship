```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\SUSHMITHA RAJU\Downloads\40811eyack.com.MAIL_xsbsxxypt8dh6!App\Dataset._csv.csv")
price_counts=df['Price range'].value_counts().sort_index()
total=len(df)
print("="*50)
print("price range ")
print("="*50)
for price,count in price_counts.items():
    percentage=(count/total)*100
    print(f"price range{price}:{count} restaurants({percentage:.2f}%)")
plt.figure(figsize=(10, 6))
plt.bar(price_counts.index, price_counts.values,
        color=['green', 'steelblue', 'hotpink', 'tomato'],
        edgecolor='black')
plt.title("Price Range Distribution of Restaurants",
          fontsize=16, fontweight='bold')
plt.xlabel("Price Range", fontsize=12)
plt.ylabel("Number of Restaurants", fontsize=12)
plt.xticks([1, 2, 3, 4],
           ['Cheap(1)', 'Moderate(2)', 'Expensive(3)', 'Very Expensive(4)'])
plt.tight_layout()
plt.show()
print("Chart saved!")
```
