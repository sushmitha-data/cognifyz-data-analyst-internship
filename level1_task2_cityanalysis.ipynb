```python
# LEVEL 1 - TASK 2: City Analysis
import pandas as pd
import matplotlib.pyplot as plt 
df = pd.read_csv(r"C:\Users\SUSHMITHA RAJU\Downloads\40811eyack.com.MAIL_xsbsxxypt8dh6!App\Dataset._csv.csv")

city_counts = df['City'].value_counts()
top_citys = city_counts.idxmax()
print("=" * 50)
print("CITY ANALYSIS FOR RESTAURANTS")
print("=" * 50)
print(f" The City with most restaurants: {top_citys} ({city_counts.max()})")
city_ratings = df.groupby('City')['Aggregate rating'].mean()
rated_city = city_ratings.idxmax()
print(f"City with highest average rating: {rated_city} ({city_ratings.max():.2f})")

plt.figure(figsize=(12, 6))
city_counts.head(10).plot(kind='bar', color=plt.cm.tab10.colors)
plt.title("Top 10 Cities by Number of Restaurants", fontsize=16, fontweight='bold')
plt.xlabel("City", fontsize=12)
plt.ylabel("Number of Restaurants", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("Level1_Task2_City_Analysis.png")
plt.show()
print("Chart saved!")
```
